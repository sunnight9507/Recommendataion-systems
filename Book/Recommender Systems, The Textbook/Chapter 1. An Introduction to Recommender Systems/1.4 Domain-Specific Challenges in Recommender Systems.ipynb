시간 데이터, 위치 기반 데이터 및 소셜 데이터와 같은 여러 도메인에서 추천 시스템은 중요한 역할을 한다. 따라서 **상황별 추천 시스템**의 개념은 이러한 **도메인에서 발생하는 추가 부가 정보를 다루기 위해 개발**되었다. 이 개념은 시간 데이터, 위치 데이터 또는 소셜 데이터와 같은 다양한 유형의 데이터에 대해 서로 다른 수정과 함께 사용된다.

# 1.4.1 Context-Based Recommender Systems

> - 상황 기반 또는 상황 인식 추천 시스템은 **다양한 유형의 상황 정보를 고려하면서 추천**을 한다.
    - 이러한 상황 정보에는 시간, 위치 또는 소셜 데이터가 포함될 수 있다.
    - 예를 들어, 소매상이 추천하는 옷의 유형은 계절과 고객의 위치에 따라 달라질 수 있다.
    - 또 다른 예는 특정 유형의 축제나 공휴일이 고객 활동에 영향을 미치는 경우다.
    - 이러한 **상황 정보의 사용**은 **추천 프로세스의 효율성을 크게 향상시킬 수 있다**는 것이 일반적으로 관찰되었다.
- **상황 기반 추천시스템은 기초적인 아이디어가 다양한 도메인별 설정과 관련이 있기 때문에 믿을 수 없을 정도로 강력하다.**

# 1.4.2 Time-Sensitive Recommender Systems

> - 많은 설정으로 항목에 대한 추천은 **`시간`에 따라 진화**할 수 있다.
    - 예를 들어, 영화에 대한 추천은 몇 년 후 받은 추천과 개봉 당시 받은 추천과 매우 다를 수 있다.
- 이러한 경우 추천 과정에 시간적 지식을 통합하는 것이 매우 중요하다.

> - 시간 추천 시스템의 시간적 측면은 여러 방식으로 반영 될 수 있다.
    - **1. 시간이 지남에 따라 사용자의 관심이 변화함에 따라 항목의 등급은 시간에 따라 진화할 수 있다.** 사용자의 관심, 호불호, 유행은 필연적으로 시간과 함께 변한다.
    - **2. 항목의 등급은 특정 시간, 요일, 월 또는 계절에 따라 달라질 수 있다.** 예를 들어, 여름 동안 겨울 의류를 추천하거나 건기 동안 우비를 추천하는 것은 말이 안 된다.
    - 첫 번째 유형의 추천 시스템은 협업 필터링 시스템에서 매개 변수로 시간을 통합하여 추천한다.
    - 두 번째 유형은 컨텍스트 기반 추천 시스템의 특수한 경우로 볼 수 있다.
    
> - **시간 추천 시스템은 등급 행렬이 희박하다는 사실 때문에 도전적**이며, 시간의 사용은 **희소성 문제를 악화**시킨다.
- 따라서 이러한 설정에서 대용량 데이터 세트 사용 여부가 특히 중요하다.

> - 웹 및 기타 인터넷 플랫폼의 사용자 활동은 향후 활동에 대한 추천을 하기 위해 사용할 수 있는 많은 유용한 데이터를 생성한다.
- 이러한 경우 이산 순차 패턴 마이닝 및 Markov 모델이 도움이 된다.
