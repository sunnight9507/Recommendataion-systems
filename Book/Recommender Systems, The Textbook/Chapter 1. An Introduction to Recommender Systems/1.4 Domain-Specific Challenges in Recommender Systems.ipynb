시간 데이터, 위치 기반 데이터 및 소셜 데이터와 같은 여러 도메인에서 추천 시스템은 중요한 역할을 한다. 따라서 **상황별 추천 시스템**의 개념은 이러한 **도메인에서 발생하는 추가 부가 정보를 다루기 위해 개발**되었다. 이 개념은 시간 데이터, 위치 데이터 또는 소셜 데이터와 같은 다양한 유형의 데이터에 대해 서로 다른 수정과 함께 사용된다.

# 1.4.1 Context-Based Recommender Systems

> - 상황 기반 또는 상황 인식 추천 시스템은 **다양한 유형의 상황 정보를 고려하면서 추천**을 한다.
    - 이러한 상황 정보에는 시간, 위치 또는 소셜 데이터가 포함될 수 있다.
    - 예를 들어, 소매상이 추천하는 옷의 유형은 계절과 고객의 위치에 따라 달라질 수 있다.
    - 또 다른 예는 특정 유형의 축제나 공휴일이 고객 활동에 영향을 미치는 경우다.
    - 이러한 **상황 정보의 사용**은 **추천 프로세스의 효율성을 크게 향상시킬 수 있다**는 것이 일반적으로 관찰되었다.
- **상황 기반 추천시스템은 기초적인 아이디어가 다양한 도메인별 설정과 관련이 있기 때문에 믿을 수 없을 정도로 강력하다.**

# 1.4.2 Time-Sensitive Recommender Systems

> - 많은 설정으로 항목에 대한 추천은 **`시간`에 따라 진화**할 수 있다.
    - 예를 들어, 영화에 대한 추천은 몇 년 후 받은 추천과 개봉 당시 받은 추천과 매우 다를 수 있다.
- 이러한 경우 추천 과정에 시간적 지식을 통합하는 것이 매우 중요하다.

시간 추천 시스템의 시간적 측면은 아래 두가지 방식으로 반영 될 수 있다.

> - **1. 시간이 지남에 따라 사용자의 관심이 변화함에 따라 항목의 등급은 시간에 따라 진화할 수 있다.** 사용자의 관심, 호불호, 유행은 필연적으로 시간과 함께 변한다.
- **2. 항목의 등급은 특정 시간, 요일, 월 또는 계절에 따라 달라질 수 있다.** 예를 들어, 여름 동안 겨울 의류를 추천하거나 건기 동안 우비를 추천하는 것은 말이 안 된다.
- 첫 번째 유형의 추천 시스템은 협업 필터링 시스템에서 매개 변수로 시간을 통합하여 추천한다.
- 두 번째 유형은 컨텍스트 기반 추천 시스템의 특수한 경우로 볼 수 있다.
    
> - **시간 추천 시스템은 등급 행렬이 희박하다는 사실 때문에 도전적**이며, 시간의 사용은 **희소성 문제를 악화**시킨다.
- 따라서 이러한 설정에서 대용량 데이터 세트 사용 여부가 특히 중요하다.

> - 웹 및 기타 인터넷 플랫폼의 사용자 활동은 향후 활동에 대한 추천을 하기 위해 사용할 수 있는 많은 유용한 데이터를 생성한다.
- 이러한 경우 이산 순차 패턴 마이닝 및 Markov 모델이 도움이 된다.

# 1.4.3 Location-Based Recommender Systems

> - **위치 기반 추천시스템은 휴대폰과 다른 GPS 가능 장치의 보급이 증가함에 따라 최근 몇 년 동안 관심이 증가하고 있다.**
    - GPS가 가능한 휴대전화의 인기가 높아지면서 소비자들이 위치기반 추천에 관심을 갖는 경우가 많다.
    - 예를 들어, 여행하는 사용자는 다른 레스토랑에 대한 자신의 이전 평가 기록을 기반으로 가장 가까운 레스토랑을 추천해주기를 원할 수 있다.
- 일반적으로 장소의 추천은 항상 그 안에 위치적인 측면이 내재되어 있다.
    - 이러한 시스템의 한 예로 레스토랑이나 유흥업소 등 다양한 형태의 장소를 추천하는 포스퀘어라는 어플이 있다.

- 이러한 시스템에 공통적인 두 가지 유형의 공간 지역성이 있다.

> - **1) User-specific locality**
- 사용자의 지리적 위치는 사용자의 선호도에 중요한 역할을 한다.
    - 예를 들어, 위스콘신 출신의 사용자는 뉴욕 출신의 사용자와 동일한 영화 선호도를 가지고 있지 않을 수 있다.
    
> - **2) Item-specific locality**
- 아이템(예를 들어, 레스토랑)의 지리적 위치는 사용자의 현재 위치에 따라 아이템의 관련성에 영향을 미칠 수 있다.
    - 사용자들은 일반적으로 현재 위치에서 매우 멀리 식사하러 갈 이유가 없다.
    
> - 위 두가지에 대한 알고리즘은 상당히 다르다.
- 전자는 상황에 맞는 시스템에 더 가깝지만 후자는 일반적으로 휴리스틱으로 설계된다.


# 1.4.4 Social Recommender Systems

> - **소셜 추천시스템은 네트워크 구조, 사회적 단서 및 태그 또는 이러한 다양한 네트워크 측면의 조합을 기반으로 한다.**
- 일반적으로 사회적 단서와 태그를 기반으로하는 소셜 추천시스템은 순전히 구조적 측면에 기반을 둔 추천시스템과 약간 다르다.
    - 순전히 구조적 측면에 기반한 추천 시스템은 네트워크 자체 내의 노드와 링크를 제안하는 데 사용된다.
    - 한편, 소셜 추천시스템도 social cue 사용이 가능한 다양한 제품을 추천하는 데 이용될 수 있다.
    
## 1.4.4.1 Structural Recommendation of Nodes and Links

> - **소셜 네트워크를 포함한 다양한 형태의 네트워크는 노드와 링크로 구성되어 있다.**
- 대부분 노드와 링크를 추천하는 것이 바람직하다.
    - 예를 들어, 개인화된 웹 검색에는 특정 주제와 관련된 자료(웹)의 추천이 필요할 수 있다.
    - 웹은 그래프로 볼 수 있기 때문에, 그러한 방법은 노드 추천 문제로 볼 수 있다.
- **노드 추천 문제는 웹 검색 문제와 밀접한 관련이 있다.**
- 이러한 방법의 핵심 구성 요소는 PageRank 알고리즘을 사용하는 것이지만 이러한 알고리즘의 개인화는 추천 알고리즘과 더 밀접하게 관련되어 있다.
- 이러한 알고리즘을 개인화된 PageLank 알고리즘이라고도 한다.
- 관심있는 노드의 예를 사용할 수 있는 경우 이러한 노드는 관심있는 다른 노드를 결정하기 위해 훈련 데이터로 사용할 수 있다.
- 밀접하게 관련된 문제는 링크 추천 또는 링크 예측 문제로, 소셜 네트워크에서 사용자에게 친구 (또는 잠재적 링크)를 제안하는 것이다.
- 링크 예측 문제는 소셜 네트워크를 넘어서 수많은 응용 프로그램을 가지고 있다.
- 흥미롭게도, 순위 결정, 집단 분류, 링크 추천의 문제들은 밀접하게 연관되어 있다.
- 사실, 한 문제에 대한 해결책은 종종 다른 문제의 하위 루틴으로 사용된다.
    - 예를 들어, 사용자 항목 그래프의 전통적인 제품 권장사항에는 종종 순위 및 링크 예측 방법이 사용된다.
    - 실제로 이러한 방법을 사용하여 여러 문제 설정에서 추천을 수행 할 수 있으며 이를 그래프로 변환 할 수 있다.
